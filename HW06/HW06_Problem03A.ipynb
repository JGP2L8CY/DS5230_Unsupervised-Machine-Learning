```python
### Import library & data set
# library
from google.colab import drive
drive.mount("/content/drive")
import re
import random
import numpy as np
import pandas as pd
import networkx as nx
from collections import defaultdict
from networkx import edge_betweenness_centrality
from networkx.algorithms.community.centrality import girvan_newman

# dataset
# node
nd_path = "/content/drive/MyDrive/DS5230_Unsupervised Machine Learning/Assignments/HW06/soc-Flickr-ASU.node_labels.txt"
node_df = pd.read_csv(nd_path,sep=",",header=None)
# node_df = node_df[:50000]
node_df = node_df[1]

# edge
eg_path = "/content/drive/MyDrive/DS5230_Unsupervised Machine Learning/Assignments/HW06/soc-Flickr-ASU.edges.txt"
edge_df = pd.read_csv(eg_path,sep=",",header=None)
# edge_df = edge_df[:50000]
edge_df = dict(zip(edge_df[0],edge_df[1]))
# edge_df = list(edge_df.items())

# to_graph
df = dict(zip(node_df,edge_df.items()))
```

```python
### edge betweeness
edge_df = {}
def edge_betweenness(graph):
  betweenness = nx.edge_betweenness_centrality(graph)
  sorted_edges = sorted(betweenness.items(),key=lambda x:x[1],reverse=True)
  return sorted_edges

# set dataset to type of graph
G = nx.Graph(df)
edge_betweenness_values = edge_betweenness(G)

# results
for edge, value in edge_betweenness_values[:]:
    edge_df[edge]=value

# make communities
approximation_groups = defaultdict(list)
approximation_threshold = 0.000000000000001

# make to group based on betweeness value
for x1, y1 in edge_df:
    approximated = False
    for group_key, group_values in approximation_groups.items():
        x2, y2 = group_key
        if abs(x1 - x2) <= approximation_threshold and abs(y1 - y2) <= approximation_threshold:
            group_values.append((x1, y1))
            approximated = True
            continue
    if not approximated:
        approximation_groups[(x1, y1)].append((x1, y1))

# result
for group_values in approximation_groups.values():
    print(group_values)
```
