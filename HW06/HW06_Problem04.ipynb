```python
### Import library & data set
# library
from google.colab import drive
drive.mount("/content/drive")
import re
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import NMF
from sklearn.decomposition import PCA
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer

# dataset
# knowledge graph
graph_path = "/content/drive/MyDrive/DS5230_Unsupervised Machine Learning/Assignments/HW06/knowledge_graph.txt"
df_graph = pd.read_csv(graph_path,header=None)
for i in range(len(df_graph)):
  df_graph[0][i]=df_graph[0][i].replace("[","").replace("'","")
  df_graph[1][i]=df_graph[1][i].replace("'","")
  df_graph[2][i]=df_graph[2][i].replace("]","").replace("'","")
df_graph.rename(columns={0:"entity",1:"content",2:"relation"},inplace=True)

# questions
questions_path = "/content/drive/MyDrive/DS5230_Unsupervised Machine Learning/Assignments/HW06/questions.txt"
df_questions = pd.read_csv(questions_path,header=None)
for i in range(len(df_questions)):
  df_questions[0][i]=df_questions[0][i].replace("[","").replace("'","")
  df_questions[1][i]=df_questions[1][i].replace("'","")
  df_questions[2][i]=df_questions[2][i].replace("]","").replace("'","")
df_questions.rename(columns={0:"question_id",1:"content",2:"entity"},inplace=True)
```

```python
### build X, y
# tokenizing X
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df_graph["content"])
# implement nmf X
nmf = NMF(n_components=1)
x_val = nmf.fit_transform(X)
# sort X
x_val = pd.DataFrame(x_val)
x_val.rename(columns={0:"val"},inplace=True)

# tokenizing y
y = vectorizer.fit_transform(df_questions["content"])
# implement nmf y
y_val = nmf.fit_transform(y)
# sort y
y_val = pd.DataFrame(y_val)
y_val.rename(columns={0:"val"},inplace=True)

# rearrange dataset
df_graph.drop(["content"],axis=1,inplace=True)
X = pd.concat([df_graph,x_val],axis=1)
# df_questions.drop(["content"],axis=1,inplace=True)
y = pd.concat([df_questions,y_val],axis=1)
```

```python
### build knowledge base
def fit(x,y):
  y_pred = []
  for i in range(len(x)):
    compare = []
    for j in range(len(y)):
      """get entity of x[i] and y, which has the cloesest val"""
      compare.append(abs(x.loc[i]["val"]-y.loc[j]["val"]))
    near_val = np.argmax(compare)

    # find entity match

    for k in range(len(y)):
      if x.loc[i]["entity"]==y.loc[j]["entity"] or x.loc[i]["relation"]==y.loc[j]["entity"]:
          y_pred.append(y.loc[j]["question_id"])
          continue

    # find the closest val
    y_pred.append(y.loc[near_val]["question_id"])
  return y_pred
# implement
y_pred = fit(X,y)

"""
========================================== result ==========================================
y_pred means that if input the train data X,
the model directs each X data to proper questions_id
"""
y_pred
```
