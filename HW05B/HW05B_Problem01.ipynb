```python
### A
# uniform
uni_nums = np.random.uniform(-1,1,1000)
count, bins, ignored = plt.hist(uni_nums, 15, density=True)
plt.show()
print("min:",min(uni_nums),"\n","max:",max(uni_nums),"\n","size:",len(uni_nums))
print("="*70)
# gaussian distribution
mu, sigma = 0, 0.1
dis = np.random.normal(mu, sigma, 1000)
count, bins, ignored = plt.hist(dis, 15, density=True)
plt.show()
print("mu:",mu,"\n","sigma:",sigma,"\n","size:",len(dis))
```

```python
### B
mu1,sigma1 = -1.0,1.0
mu2,sigma2 = 1.5,1.0
x = np.linspace(-5,5,1000)
y1 =  (1 / np.sqrt(2 * np.pi * sigma1**2)) * np.exp(-(x-mu1)**2 / (2 * sigma1**2))
y2 = (1 / np.sqrt(2 * np.pi * sigma2**2)) * np.exp(-(x-mu2)**2 / (2 * sigma2**2))
plt.plot(x, y1, alpha=0.7, label=r'PDF of N(0, $1^2$)')
plt.plot(x, y2, alpha=0.7, label=r'PDF of N(1.5, $1.5^2$)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
print("mu1,mu2:",mu1,mu2,"\n","sigma1,sigma2:",sigma1,sigma2,"\n","size:",1000)
```

```python
### C
proba = [i for i in range(1,401)]
probas = np.random.dirichlet(proba,size=1)
probas = np.concatenate(probas).tolist()
samples_1 = np.random.choice(len(proba),20,replace=False,p=probas)
count, bins, ignored = plt.hist(samples_1, 15, density=True)
plt.show()
print(samples_1)
```
